```python
import re

def check_birth_year(detail):
  return len(detail) == 4 and int(detail) >= 1920 and int(detail) <= 2002

def check_issue_year(detail):
  return len(detail) == 4 and int(detail) >= 2010 and int(detail) <= 2020

def check_expiration_year(detail):
  return len(detail) == 4 and int(detail) >= 2020 and int(detail) <= 2030

def check_height(detail):
  if detail.endswith('cm'):
    detail = detail.replace('cm', '')
    return int(detail) >= 150 and int(detail) <= 193 
  elif detail.endswith('in'):
    detail = detail.replace('in', '')
    return int(detail) >= 59 and int(detail) <= 76

def check_hair_color(detail):
  return re.search(r'#[0-9a-f]{6}', detail) is not None

def check_eye_color(detail):
  return detail in [ 'amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth' ]

def check_passport(detail):
  # return len(detail) == 9 and detail.isnumeric()
  return re.search(r'[0-9]{9}', detail) is not None

def check_any_document(doc, detail):
  if doc == 'byr':
    return check_birth_year(detail)
  elif doc == 'iyr':
    return check_issue_year(detail)
  elif doc == 'eyr':
    return check_expiration_year(detail)
  elif doc == 'hgt':
    return check_height(detail)
  elif doc == 'hcl':
    return check_hair_color(detail)
  elif doc == 'ecl':
    return check_eye_color(detail)
  elif doc == 'pid':
    return check_passport(detail)
```

```python
valid_passports_w_deets = 0
for passport in passport_docs_deets:
  document_check = [check_any_document(doc, detail) for doc, detail in zip(passport.keys(), passport.values()) if doc != 'cid']
  print(f'document_check: {document_check}')
  if all(document_check):
    valid_passports_w_deets += 1
```
