```python
with open('input.txt', 'r') as infile:
  groups = infile.read().split('\n\n')
```

```python
group_answers = [set(answers.replace('\n', '')) for answers in groups]

# for group in group_answers:
#   print(f'group: {group}')

print(f'sum unique answers: {sum([len(answers) for answers in group_answers])}')
```

```python
groups_shared_answers = []

for group in groups:
  group_shared_answers = set(group.split('\n')[0])
  if len(group.split('\n')) > 1:
    for person in group.split('\n')[1:]:
      group_shared_answers= group_shared_answers.intersection(set(person))
  groups_shared_answers.append(group_shared_answers)
  
print(f'groups_shared_answers: {groups_shared_answers}')
print(f'sum of shared answers: {sum([len(answers) for answers in groups_shared_answers])}')
```
